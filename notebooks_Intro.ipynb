```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K = 1
a = 0.5
z = np.exp(1j * np.linspace(0, 2 * np.pi, 500))
H = K / (1 + a / z)
```

```python
# Plot the magnitude response
plt.plot(np.real(z), np.abs(H))
plt.title("Magnitude Response")
plt.xlabel("Real")
plt.ylabel("Absolute Value")
plt.grid()
plt.show()
```

```python
# Plot the phase response
plt.plot(np.real(z), np.angle(H))
plt.title("Phase Response")
plt.xlabel("Real")
plt.ylabel("Phase (Radians)")
plt.grid()
plt.show()
```

```python
# Plot the pole-zero diagram
plt.scatter(-1/a, 0, color='red')
plt.scatter(0, 0, color='black')
plt.title("Pole-Zero Diagram")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.grid()
plt.show()
```

```python
# Compute the Fourier transform
from scipy.fftpack import fft
w, H_fourier = np.linspace(0, 2 * np.pi, 500, retstep=True)
H_fourier = np.abs(fft(H)) * H_fourier
```

```python
# Plot the Fourier transform
plt.plot(w, H_fourier)
plt.title("Fourier Transform")
plt.xlabel("Frequency (Radians)")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```
